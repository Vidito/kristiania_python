```python
def shared_characters(str1, str2):
    # Characters that appear in both strings
    set1 = set(str1)
    set2 = set(str2)
    shared = set1.intersection(set2)
    return sorted(shared)

def unique_characters(str1, str2):
    # Characters that only appear in one of the strings
    set1 = set(str1)
    set2 = set(str2)
    unique = set1.symmetric_difference(set2)
    return sorted(unique)

def non_occurring_letters(str1, str2):
    # Characters from a-z that are not in either string
    alphabet = set('abcdefghijklmnopqrstuvwxyz')
    set1 = set(str1.lower())
    set2 = set(str2.lower())
    combined = set1.union(set2)
    non_occurring = alphabet.difference(combined)
    return sorted(non_occurring)


str1 = input("Enter the first string: ")
str2 = input("Enter the second string: ")

print("\nShared characters:")
print(', '.join(shared_characters(str1, str2)))

print("\nUnique characters")
print(', '.join(unique_characters(str1, str2)))

print("\nNon-occurring alphabet letters:")
print(', '.join(non_occurring_letters(str1, str2)))
```

```python
# Creating a class for Person
class Person:
    """A base class with basic information"""

    def __init__(self, fname, lname, age):
        self.fname = fname
        self.lname = lname
        self.age = age

    def get_info(self):
        print("Full Name:", self.fname, self.lname)
        print("Age:", self.age)


class Student(Person):
    """Student class inherits from Person and adds student ID"""

    def __init__(self, fname, lname, age, student_ID):
        # Super() to use Persons attributes
        super().__init__(fname, lname, age)
        self.student_ID = student_ID # Student spesific attribute

    def get_stuinfo(self):
        """Print student information including inherited attributes"""
        self.get_info()
        print("Student ID:", self.student_ID)


class Employee(Person):
    """Employee class inherits from the Person class and then adds employee number and salary"""

    def __init__(self, fname, lname, age, employee_number, salary):
        super().__init__(fname, lname, age) #Reusing Person's attributes
        self.employee_number = employee_number # Attributes for the employees
        self.salary = salary

    def get_empinfo(self):
        """Printing employee information with its inherited attributes"""
        self.get_info() # Using Person's method
        print("Employee No:", self.employee_number)
        print("Salary:", self.salary, "USD")


# Testing the code
new_student = Student("Anthony", "Smith", 35, "s346571")
new_student.get_stuinfo()
print("="*30)
new_employee = Employee("Sarah", "Taylor", 34, 2919736, 5000)
new_employee.get_empinfo()
```
