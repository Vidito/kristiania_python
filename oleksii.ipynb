```python
#This program asks the user for three file names: one containing regular text
#and one containing a list of sensitive words.
#It then replaces all occurrences of the sensitive words in the text
#with asterisks and saves the result to a third file,
#where the sensitive words are concealed.

#the program allows the user to specify the next:
text_file_name = input("Enter the name of the original text file: ")
words_file_name = input("Enter the name of the sensitive words file: ")
redacted_file_name = input("Enter the name of the redacted file: ")

#reads the original text file
# first we open the file in read mode
orig_textfile = open(text_file_name, "r")
# then we save the content in a variable
txt = orig_textfile.read()
# we make sure we close the file
orig_textfile.close()

#reads the sensitive words file
s_words = open(words_file_name, "r")
s_words_text = s_words.read()
s_words.close()

#replaces sensitive words
# split the text by \n since words are on separate lines
for w in s_words_text.split("\n"):
  # if the length of the word is 3 letters, it will be replaced by 3 stars
  txt = txt.replace(w,"*" * len(w))

#saves the redacted version of the original text in a new file.
# we use write mode to create a new file or overwrite the existing one
redacted_version = open(redacted_file_name, "w")
# we write the redacted text processed before into the opened file
redacted_version.write(txt)
redacted_version.close()
```
